```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('/home/haoyang-22/project/event-camera/ultralytics/detect_0301/runs/segment/train5/weights/best.pt')

# Open the video file
video_path = "/home/haoyang-22/project/event-camera/0225_detectweight/f_100g_recording_2024-02-25_14-35-36 00_06_20-00_06_44.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model.predict(frame,line_width=1)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
#Qmodel = YOLO('/home/haoyang-22/project/event-camera/ultralytics/detect_0301/yolov8n.pt')
model = YOLO('/home/haoyang-22/project/event-camera/ultralytics/detect_0301/yolov8n-seg.pt')
#model = YOLO('/home/haoyang-22/project/event-camera/ultralytics/detect_0301/yolov8n-pose.pt')


# Open the video file
video_path = "/home/haoyang-22/project/event-camera/ultralytics/detect_0301/100 Different Ways to Walk by Kevin Parry (Animation Reference with Music).mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.track(frame, persist=True)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Tracking", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
